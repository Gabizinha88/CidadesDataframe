# ** **negrito**1- Apresente em tela (output) toda a base de dados. Pontuação: 0,5 pontos**

````python
import pandas as pd

# começando meu dataframe baseado em cidades de amigos
data = {
    'Nome': ['Clebler', 'Michael', 'Maria Lua', 'Junior Largatixa', 'Tarik Jr', 'Yasmin'],
    'Idade': [23, 34, 10, 36, 13, 27],
    'Cidade': ['Jardim de Piranhas', 'Xique Xique', 'Fercal', 'New York', 'Rio de Janeiro', 'London'],
    'Estado': ['RN', 'BA', 'DF', 'NY', 'RJ', 'ENG'],
    'CEP': ['59324-000', '44900-000', '73255-000', '10001', '20000-000', 'WC2N 5DU'],
    'País': ['Brasil', 'Brasil', 'Brasil', 'Estados Unidos', 'Brasil', 'Reino Unido']
}

# importando os dados
cidades = pd.DataFrame(data)

# 1. apresentar a saida completa
print("Base de dados completa:")
print(cidades)
````

# *2- Apresente o tamanho do seu dataframe (quantas colunas x linhas).*

````python
num_linhas, num_colunas = cidades.shape
print(f"O DataFrame tem {num_linhas} linhas e {num_colunas} colunas.")
````

#3- **negrito** Acesse a linha (x) e apresente em tela todas as características do item.

````python
cidades = pd.DataFrame(data)

# Vamos definir a linha que queremos acessar e seu indice
x = 2

# Olhando se o indice se encontra no mesmo intervalo
if 0 <= x < len(cidades):
    # Acessando a linha x e exibindo todas as características
    print(f"Características do item na linha {x}:")
    print(cidades.iloc[x])
else:
    print(f"O índice {x} está fora do intervalo do DataFrame.")
````

**4- Verifique se o dataframe está vazio. Pontuação: 0,5 pontos**


````python
is_empty = cidades.empty

# 2. Verificar se o DataFrame está vazio ou não
if is_empty:
    print("\nO DataFrame está vazio.")
else:
    print("\nO DataFrame não está vazio.")

````

**7- Adicione um item (linha) na sua base de dados. Pontuação: 0,5 pontos**

````python
# Adicionar uma nova linha
new_row = {
    'Nome': 'Alice',
    'Idade': 29,
    'Cidade': 'Paris',
    'Estado': 'Île-de-France',
    'CEP': '75000',
    'País': 'França'
}

cidades.loc[len(cidades)] = new_row

print("\nBase de dados após adicionar uma nova linha:")
print(cidades)

````
